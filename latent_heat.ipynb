## 1. ゲストが気体の場合

包接水和物内と平衡にあるゲストが、外界では気体である場合。理想気体を想定する。

### 1.1 単原子分子

単原子理想気体の化学ポテンシャルは、自由エネルギーの式

$$
f=-kT\ln\left[(mkT/2\pi\hbar^2)^\frac{3}{2}\int_V\exp\left(-\beta w({\bf r})\right){\rm d}{\bf r}\right]\tag{1.1}
$$

に、$w({\bf r})=0$(理想気体の条件、分子は相互作用しない)を入れると、[^1]

$$
f=-kT\ln\left[(mkT/2\pi\hbar^2)^\frac{3}{2}V\right]\tag{1.2}
$$

が得られ、これに$V = N kT/p$ を入れ、分子数$N$で割ると、

<!-- $$\label{eq:mu_g} -->

$$
\mu_g(T,p)=-kT\ln\left(\left(\frac{m}{2\pi\beta\hbar^2}\right)^\frac{3}{2}\frac{kT}{p}\right)\tag{1.3}
$$

となる。[^2]

これを温度で微分すると、エントロピーが得られる。

<!-- $$\label{eq:s_id} -->

$$
\frac{\partial\mu_g}{\partial T}=\mu_g/T-\frac{5}{2}k
\tag{1.4}
$$

### 1.2 直線分子

同様に、直線分子の場合には、

$$
f_a=-kT\ln\left[\sigma^{-1}\left(\frac{m}{2\pi\beta\hbar^2}\right)^\frac{3}{2}\frac{I}{\beta\hbar^2}\int_v\int_\Omega\exp\left(-\beta w({\bf r},{\bf \Omega})\right){\rm d}{\bf r}{\rm d}{\bf \Omega}\right]
\tag{1.5}
$$

から出発する。

### 1.3 立体分子

$$
f_a=-kT\ln\left[
\frac{1}{2}
\prod_{j=1}^3  \left(\frac{I_j}{2\pi\beta\hbar^2}\right)^\frac{1}{2}
\left(\frac{m}{2\pi\beta\hbar^2}\right)^\frac{3}{2}
\int_V\int_\Omega\exp\left(-\beta w({\bf r},{\bf \Omega})\right)
{\rm d}{\bf r}{\rm d}{\bf \Omega}
\right]
\tag{1.6}
$$

---

いずれも相互作用を 0 とみなして積分を$8\pi^2v$($v$は 1 分子あたりの体積)におきかえ、さらに理想気体の体積
$v = kT/p$を代入すると、 自由分子の化学ポテンシャルが得られる。

$$
% \label{eq:mu_ig}
\mu_g=-kT\ln\left[
\frac{1}{2}
\prod_{j=1}^3  \left(\frac{I_j}{2\pi\beta\hbar^2}\right)^\frac{1}{2}
\left(\frac{m}{2\pi\beta\hbar^2}\right)^\frac{3}{2}
\frac{8\pi^2}{\beta p}
\right]
\tag{1.7}
$$

温度で微分すると、エントロピーが得られる。

$$
\begin{aligned}
% \label{eq:s_id2}
S_g=\frac{\partial\mu_g}{\partial T}&=\mu_g/T-\frac{7k}{2}&(直線分子)\\
                                           &=\mu_g/T-4k&(立体分子)
\end{aligned}
\tag{1.8}
$$

[^1]: 久保(2.74)
[^2]: 久保(4.96)式に一致。


## 2. ゲスト分子が水と混じらない液体の場合

ゲスト液体はその温度での飽和蒸気圧の気体と平衡状態にあると考え、飽和蒸気圧の気体の化学ポテンシャルを上の式(1.3)や(1.7)で求め、それを液体の化学ポテンシャルとする。飽和蒸気圧は実験値を参照する。


## 3. エンタルピー

ハイドレート結晶のエンタルピーは式(0.1)の分配関数の対数を$\beta$で微分して得られる。

$$
H=-\frac{\partial\ln\Xi}{\partial\beta}=A_w^0-\sum_kN_k\frac{\sum_j(\frac{\partial(\beta\mu_j)}{\partial\beta}-\frac{\partial(\beta f_j^k)}{\partial\beta})\exp\left(\beta(\mu_j-f_j^k)\right)}{1+\exp\left(\beta(\mu_j-f_j^k)\right)}
\tag{3.1}
$$

ゲストが一種類だけの場合は、$j$での和を外し、$\mu_j\rightarrow\mu_g$、$f_j^k\rightarrow f_k$と書きなおすことにする。$k$ケージの占有率$y_k$を次のように定義する。

$$
y_k=\frac{\exp\left(\beta(\mu_g-f_k)\right)}{1+\exp\left(\beta(\mu_g-f_k)\right)}
\tag{3.2}
$$

すると上の式は簡単になり、

$$
H=-\frac{\partial\ln\Xi}{\partial\beta}=A_w^0-\sum_k N_k y_k\left(\frac{\partial(\beta\mu_g)}{\partial\beta}-\frac{\partial(\beta f_k)}{\partial\beta}\right)
\tag{3.3}
$$

水分子数$N_w$で割ると、水 1 分子あたりのエンタルピー$h$は次のようになる。

$$
% \label{eq:enthalpy}
h=h_c^0-\sum_k \alpha_k y_k\left(\frac{\partial(\beta\mu_g)}{\partial\beta}-\frac{\partial(\beta f_k)}{\partial\beta}\right)
\tag{3.4}
$$

$\mu_g$は分解状態でのゲストの(ゲスト 1 モルあたりの)化学ポテンシャルで、その温度微分は分解状態でのエントロピーになる。

$$
\begin{aligned}
\frac{\partial\mu_g}{\partial\beta} &=&\frac{\partial T}{\partial \beta}\frac{\partial\mu_g}{\partial T}\\
&=&\frac{T}{\beta}\frac{\partial\mu_g}{\partial T}
\end{aligned}
\tag{3.5}
$$

つまり、

$$
\begin{aligned}
\frac{\partial(\beta\mu_g)}{\partial\beta}&=&\mu_g+\beta\frac{\partial\mu_g}{\partial\beta}\\
&=&\mu_g-T\frac{\partial\mu_g}{\partial T}\\
&=&\mu_g-TS\\
&=&h_g
\end{aligned}
\tag{3.6}
$$

$h_g$は分解状態でのゲストの(ゲスト 1 モルあたりの)エンタルピー。(1.4), (1.8)より、[^3]

$$
\begin{aligned}
h_g&=\frac{5kT}{2}&(単原子分子)\\
&=\frac{7kT}{2}&(直線分子)\\
&=4kT&(立体分子)
\end{aligned}
\tag{3.7}
$$

次に$\beta f_k$を$\beta$で微分する。(0.4)の各項について考える。 $f_0$は積分の中核部分。

$$
\beta f_0=-\ln\int_{v_k}\int\exp\left[-\beta w({\bf r},{\bf \Omega})\right]
{\rm d}{\bf r}{\rm d}{\bf \Omega}
\tag{3.8}
$$

これを$\beta$で微分すると、

$$
\begin{aligned}
\frac{\partial \beta f_0}{\partial\beta}&=&\frac{\int_{v_k}\int w_k({\bf r},{\bf \Omega})\exp\left[-\beta w_k({\bf r},{\bf \Omega})\right]
{\rm d}{\bf r}{\rm d}{\bf \Omega}}{\int_{v_k}\int\exp\left[-\beta w_k({\bf r},{\bf \Omega})\right]
{\rm d}{\bf r}{\rm d}{\bf \Omega}}\\
&=&\bar{w}_k
\end{aligned}
\tag{3.9}
$$

$k$ケージ内の平均ポテンシャル$\bar{w}_k$を計算することになる。
質量項$a$は(単原子分子・多原子分子共通)、

$$
\begin{aligned}
\frac{\partial(\beta a)}{\partial\beta}&=&\frac{3}{2}\frac{\partial \ln\frac{2\pi\hbar^2\beta}{m}}{\partial\beta}\\
&=&\frac{3}{2}\frac{\frac{2\pi\hbar^2}{m}}{\frac{2\pi\hbar^2\beta}{m}}\\
&=&\frac{3kT}{2}
\end{aligned}
\tag{3.10}
$$

最後に慣性モーメント項$c$。(直線分子と立体分子のみ)。

直線分子の場合

$$
\begin{aligned}
\frac{\partial(\beta c)}{\partial\beta}&=&\frac{\partial \ln\frac{2\pi\hbar^2\beta}{I}}{\partial\beta}\\
&=&\frac{\frac{2\pi\hbar^2}{I}}{\frac{2\pi\hbar^2\beta}{I}}\\
&=&kT
\end{aligned}
\tag{3.11}
$$

剛体分子の場合

$$
\begin{aligned}
\frac{\partial(\beta c)}{\partial\beta}&=&\frac{1}{2}\sum_{j=1}^3\frac{\partial \ln\frac{2\pi\hbar^2\beta}{I_j}}{\partial\beta}\\
&=&\frac{1}{2}\sum_{j=1}^3\frac{\frac{2\pi\hbar^2}{I_j}}{\frac{2\pi\hbar^2\beta}{I_j}}\\
&=&\frac{3kT}{2}
\end{aligned}
\tag{3.12}
$$

$s$は分子の対称性からの寄与で、これは$\beta$をかけて$\beta$で微分すると消える。

結局、

$$
\begin{aligned}
\frac{\partial(\beta f_k)}{\partial \beta}&=\bar{w}_k+\frac{3kT}{2}&(単原子分子)\\
&=\bar{w}_k+\frac{5kT}{2}&(直線分子)\\
&=\bar{w}_k+3kT&(立体分子)
\end{aligned}
\tag{3.13}
$$

これらを(3.4)に代入し、水の分子数 $N_w$で割ると、最終的に得られる Filled hydrate のエンタルピー$h_f$の式は以下のような形になる(単原子分子の場合)

$$
% \label{eq:enthalpy2}
h_f=h_c^0-\sum_k\alpha_k y_k\left(h_g-\bar{w}_k-\frac{3kT}{2}\right)
\tag{3.14}
$$

(3.14)の右辺第 1 項は空のクラスレートのエンタルピーである。
$$h_c^0=\mu_c^0-T\frac{\partial\mu_c^0}{\partial T}$$
$\mu_c^0$はすでに計算しているので、これの温度による数値微分から$h_c^0$が求められる。
(3.14)の右辺第 2 項はゲストが占有することによって生じる項で、$\alpha_k$は結晶構造から決まる定数、$y_k$は占有率なので vdWP で推定できるが、大ケージのみを占有するクラスレートではほぼ 1 であり、計算上は省略できる。$\bar{w}_k$は平均ポテンシャルで、これは$f_k$を計算する副作用で容易に計算できる。

分解後のエンタルピーについて検討する。まず、現在扱っている条件($T>0^\circ\rm C$)では分解すると水が生じる。水のエンタルピーは、$0^\circ\rm C$での氷と水のエンタルピー差(融解潜熱)と、水の定圧比熱の温度積分(これは Antoine の式から出せるはず。)で求める。ゲストはとりあえずガスになると考える。水 1 分子に対し、ゲストの個数は$\sum_k\alpha_k y_k$、なのでその分解後のエンタルピーは$\sum_k\alpha_k y_k h_g$となるはず。

分解後のエンタルピー$h_d$は水のエンタルピーである。

$$
\begin{aligned}
h_d&=&h_w\\
&=&h_i(T=273.15)+\Delta h_m+\int_{273.15}^TC_p{\rm d}T
\end{aligned}
\tag{3.15}
$$

$\Delta h_m=h_i-h_w$は氷の融解潜熱(80 cal/g)。$C_p$は液体の水の比熱(1 cal/g/K)。水の比熱は 0℃ から 100℃ までほとんど変化しないので、実験値を使うのが良さそうだ。結局、ハイドレートの分解時の潜熱$\Delta h$は、

$$
\begin{aligned}
\Delta h&=&h_d-h_f\\
&=&h_w-\left[h_c^0-\sum_k\alpha_k y_k\left(h_g-\bar{w}_k-\frac{3kT}{2}\right)\right]\\
&=&h_i(T=273.15)+\Delta h_m+\int_{273.15}^TC_p{\rm d}T-h_c^0+\sum_k\alpha_k y_k\left(h_g-\bar{w}_k-\frac{3kT}{2}\right)
\end{aligned}
\tag{3.16}
$$
